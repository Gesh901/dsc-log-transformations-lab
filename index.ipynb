### Plot Log Transformed Versions of Features

For each feature that you identified as a good candidate for log transformation, plot the feature vs. `SalePrice` as well as the log transformed feature vs. log transformed `SalePrice`.

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

y_raw = ames["SalePrice"]
X_raw = ames["GarageArea"]
y_log = np.log(y_raw)  
X_log = np.log(X_raw)

fig,ax = plt.subplots(ncols =2, figsize =(15,8))

ax[0].scatter(X_raw,y_raw)
ax[0].set_title("Scatter Plot of SalePrice vs Garage Area")
ax[0].set_xlabel("Garage Area")
ax[0].set_ylabel("Sale Price")
ax[1].scatter(X_log,y_log)
ax[1].set_title("Scatter Plot of log(SalePrice) vs log(Garage Area)")
ax[1].set_xlabel("Log (Garage Area)")
ax[1].set_ylabel("Log(Sale Price)");



```

```python
y_raw = ames["SalePrice"]
X_raw = ames["2ndFlrSF"]
y_log = np.log(y_raw)  
X_log = np.log(X_raw)

fig,ax = plt.subplots(ncols =2, figsize =(15,8))

ax[0].scatter(X_raw,y_raw)
ax[0].set_title("Scatter Plot of SalePrice vs 2ndFlrSF")
ax[0].set_xlabel("2ndFlrSF")
ax[0].set_ylabel("Sale Price")
ax[1].scatter(X_log,y_log)
ax[1].set_title("Scatter Plot of log(SalePrice) vs log(2ndFlrSF)")
ax[1].set_xlabel("Log (2ndFlrSF)")
ax[1].set_ylabel("Log(Sale Price)");
```

## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

```python
# Your code here - prepare data for modeling
X_all = ames[["GarageArea", "2ndFlrSF", "GrLivArea"]]
X_all_log =np.log(X_all+1)
X_all_log
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model

import statsmodels.api as sm

log_model = sm.OLS(y_log,sm.add_constant(X_all_log))
model_results = log_model.fit()
```

### Model Evaluation and Interpretation

How did the model perform? How might we interpret its coefficients? Create as many cells as needed.

```python
# Your code here - evaluate the model

print(model_results.summary())
```
